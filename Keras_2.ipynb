

Example data:

An experiemental drug was tested on individuals from ages 13 to 100.
The trial had 2100 participants. Half were under 65 years old, half were  over 65 years old.
95% of patientes 65 or older experienced side effects.
95% of patients under 65 experienced no side effects.

 Neural Network Predictions with TensorFlow's Keras API

```python
import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
```

```python
train_labels = []
train_samples = []
```

```python
for i in range(50):
  # The 5% of younger individuals who did experience side effects
  random_younger = randint(13,64)
  train_samples.append(random_younger)
  train_labels.append(1)

  # The 5% of older individuals who did not experience side effects
  random_older = randint(65,100)
  train_samples.append(random_older)
  train_labels.append(0)

for i in range(1000):
  # The 95% of younger individuals who did not experience side effects
  random_younger = randint(13,64)
  train_samples.append(random_younger)
  train_labels.append(0)

  # The 95% of older individuals who did experience side effects
  random_older = randint(65,100)
  train_samples.append(random_older)
  train_labels.append(1)
```

```python
train_labels = np.array(train_labels)
train_samples = np.array(train_samples)
train_labels,train_samples = shuffle(train_labels,train_samples)
```

```python
scaler = MinMaxScaler(feature_range = (0,1))
sc_trained_samples = scaler.fit_transform(train_samples.reshape(-1,1))
```

```python
sc_trained_samples = np.array(sc_trained_samples)
sc_trained_samples
```

```python
import keras
from keras import backend as K
from keras.models import Sequential
from keras.layers import Activation
from keras.layers.core import Dense
from keras.optimizers import adam_v2
from keras.metrics import categorical_crossentropy
from keras import optimizers
from keras import metrics
```

```python
model = Sequential([
                    Dense(units = 160,input_shape = (1,),activation = 'relu'),
                    Dense(units = 320,activation = 'relu'),
                    Dense(units = 2,activation = 'softmax')
])
```

Predict

```python
prediction = model.predict(x = sc_trained_samples,batch_size = 10,verbose = 0)
```

```python
for i in prediction:
  print(i)
```

```python
rounded_prediction = np.argmax(prediction,axis = 1)
rounded_prediction
```

```python
for i in rounded_prediction:
  print(i)
```

Confusion Matrix

```python
%matplotlib inline
from sklearn.metrics import confusion_matrix
import itertools
import matplotlib.pyplot as plt
```

```python
cm = confusion_matrix(y_true = train_labels,y_pred = rounded_prediction)
```
